# Information Extraction Implementation (NER, EL)

```python
def get_parts(bird_token):
    return (bird_token.pos_ == "NOUN") and (str(bird_token) not in COLOUR_LIST) and (str(bird_token) not in SHAPE_LIST) and (str(bird_token) not in SIZE_LIST)
def get_adj(bird_token): 
    return (bird_token.pos_ == "ADJ") or (str(bird_token) in COLOUR_LIST) or (str(bird_token) in SHAPE_LIST) or (str(bird_token) in SIZE_LIST)
def get_uberon(bird_token):
    knows_query = f"""
        PREFIX obo-in-owl: <http://www.geneontology.org/formats/oboInOwl#>
        SELECT DISTINCT ?s ?l
        WHERE {{
            ?s <http://www.geneontology.org/formats/oboInOwl#hasOBONamespace> "uberon" .
            ?s rdfs:label | obo-in-owl:hasRelatedSynonym | obo-in-owl:hasExactSynonym "{bird_token}" .
            ?s rdfs:subClassOf* <http://purl.obolibrary.org/obo/CARO_0000000> .
            ?s rdfs:label ?l .
            FILTER NOT EXISTS {{?a ?s ?b}}
        }}"""
    qres = graph.query(knows_query)
    if len(qres) == 0:
        return ("No Code Found",bird_token)
    for row in qres: 
        code = (row.s).split("/")[-1]
        return(row.s, row.l)
def get_pato(attribute):
    pattern = re.compile('[Aa-zZ]*ish$') 
    if bool(re.match(pattern, str(attribute))):  
        attribute = nlp(str(attribute)[:-3])[0]
    knows_query = f"""
                PREFIX obo-in-owl: <http://www.geneontology.org/formats/oboInOwl#>
                SELECT DISTINCT ?s ?c ?d ?l
                WHERE {{
                     ?s rdfs:label "{attribute.lemma_}"^^xsd:string .
                     ?s rdfs:label ?l .
                     ?s rdfs:subClassOf ?c .
                     ?s rdfs:subClassOf* <http://purl.obolibrary.org/obo/PATO_0001241> .
                     ?c rdfs:label ?d .
                }}"""
    qres = graph_pato.query(knows_query)
    if len(qres) == 0:
        knows1_query = f"""
                PREFIX obo-in-owl: <http://www.geneontology.org/formats/oboInOwl#>
                SELECT DISTINCT ?s ?c ?d ?l
                WHERE {{
                     ?s obo-in-owl:hasExactSynonym | obo-in-owl:hasRelatedSynonym "{attribute.lemma_}"^^xsd:string .
                     ?s rdfs:label ?l .
                     ?s rdfs:subClassOf ?c .
                     ?s rdfs:subClassOf* <http://purl.obolibrary.org/obo/PATO_0001241> .
                     ?c rdfs:label ?d .
                }}"""
        qres = graph_pato.query(knows1_query)
    if len(qres) == 0:
        return "No Pato Found"
    for row in qres:
        return((row.s,row.c,row.l,row.d)) 
```

```python
BODY = get_uberon("body")
complete_bird_dict = {} 
counter = 0
for i in range(len(df["Description"])):
    description = df["Description"][i]
    doc = nlp(description)
    matches = matcher(doc) #extract information
    bird_dict = {BODY:[]}
    for x in range(len(matches)):
        noun = BODY
        match_id, token_ids = matches[x]
        p_list = []
        for z in range(len(token_ids)):  #NER and EL
            sorted_ids = sorted(token_ids)
            if (sorted_ids[-1] - sorted_ids[0]) > 8:
                continue  
            bird_token = doc[token_ids[z]]
            if get_parts(bird_token):
                noun = get_uberon(bird_token.lemma_)
                if noun not in bird_dict.keys() and noun[0] != "No Code Found":
                    bird_dict[noun] = []
            elif get_adj(bird_token):
                code = get_pato(bird_token)
                if noun[0] != "No Code Found" and code not in bird_dict[noun] and code != 'No Pato Found':
                    p_list.append(code)  
            if z == len(token_ids)-1 and noun[0] != "No Code Found":
                bird_dict[noun] = list(set(bird_dict[noun] + p_list))
    bird_name = df["Bird Name"][i]
    for k,v in bird_dict.copy().items():
        if len(v) == 0:
           del bird_dict[k]
    complete_bird_dict[bird_name] = bird_dict
    counter += 1 
    if counter % 5 == 0:
        print("progress: ",counter,"/200")
print(complete_bird_dict)
```

# Knowledge Graph Construction

```python
def transform_code(code):
    return f"http://purl.obolibrary.org/obo/{code}"
def transform_bird_name(name):
    name = name.replace(" ","")
    return f"http://example.org/{name}"
```

```python
from rdflib import RDFS
cg = Graph()
has = URIRef("http://example.org/has")
#for all birds 
for bird in complete_bird_dict:
    bird_iri = transform_bird_name(bird)
    name = URIRef(bird_iri)
    for part in complete_bird_dict[bird]:
        ub_part,ub_label = part[0],part[1]
        for pato in complete_bird_dict[bird][part]:
            if pato == "No Pato Found":
                continue
            pato_code,class_code,pato_label,class_label = pato[0],pato[1],pato[2],pato[3]
            if (pato_code == ' '): 
                continue
            cg.add((name, ub_part, pato_code))
            cg.add((ub_part, RDFS.label,ub_label))
            cg.add((pato_code,RDFS.label,Literal(pato_label)))
            cg.add((pato_code, RDF.type, class_code))
            cg.add((class_code, RDFS.label, Literal(class_label)))
for s, p, o in cg:
    print((s, p, o)) 
cg.serialize(destination = "all_type_labelgraph_31jul.nt",format="ntriples")
```
